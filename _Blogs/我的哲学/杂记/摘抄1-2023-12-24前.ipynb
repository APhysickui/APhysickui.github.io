深夜平躺在床上，意识到自己陷入了结构性困境，意识到自己的劳动毫无价值，意识到虚无主义在埋葬你本就灰暗的未来，意识到活着就像一种罪，像发酵的臭垃圾，像空中飞舞的垃圾袋突然就想到哲学家笑话，《有人陷入了虚无主义症状，有没有哲学博士帮帮忙!“别担心先生，我思故我在”》。可惜哲学家都死了，哲学也死了，只有你还活着。而活着就需要论文，需要绩点，需要社交。他们围着你问:“躺够了吗，这个期刊是……，我们来讨论下……，创新点在……”。他们走的时候说:“兼容并包，北大会容许你这样失败者的存在。” 是啊，失败者。可是你走了好远，才成为了偏离路线的失败者，你好想告诉他们，我这样活着只是一种自古以来的生存困境，但就像在荒野里独行，开口说话只是更加寂寞。好寂寞啊，没有邮件，没有微信，什么都没有，自由得像一只游荡的幽灵，你忽然想到庄子说，我怎么知道生不是一场梦，死不是游子害怕返回故乡呢？开玩笑的，你只是会憎恨这一切，你学过的概念没有庇护你，你的痛苦却被分门别类地包装好，成为医院预约名单的长龙之一。真有趣啊，你想。整天希望逃避人群的你，在咨询室里，竟然要向这个世界乞讨，希望获得一丝微不足道的谅解，并且还要像个销售一样推销自己，希望被世界认同。惶惶不可终日。
