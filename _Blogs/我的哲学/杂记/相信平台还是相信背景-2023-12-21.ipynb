> 楚襄王问于宋玉曰:“先生其有遗行与? 何士民众庶不誉之甚也!<br/>
> 宋玉对日:“唯，然，有之!愿大王宽其罪，使得毕其辞。客有歌于郢中者，其始日《下里》《巴人》，国中属而和者数千人。其为《阳阿》、《瘫露》，国中属而和者数百人。其为《阳春》、《白雪》，国中有属而和者，不过数十人。引商刻羽，杂以流徽，国中属而和者，不过数人而已。是其曲弥高，其和弥寡。<br/>
> 故乌有凤而鱼有鳃。凤凰上击九千里，绝云霓，负苍天，足乱浮云，翱翔乎杳冥之上。夫蕃篱之舅，岂能与之料天地之高哉?鳃鱼朝发昆仑之墟，暴警于碣石，暮宿于孟诸。夫尺泽之貌，岂能与之量江海之大哉?故非独乌有凤而鱼有鳃，士亦有之。夫圣人瑰意琦行，超然独处，世俗之民，又安知臣之所为哉?

宋玉《对楚王问》

我有时候也在怀疑是不是走错了路，因为在安徽省只有省委的公务员会有这些偏见（或者说文化），我在北大没有看到这些，我也不知道这些明明是最优秀的一批年轻人以后会往哪里走，未来到底是不是他们的？

但我没有回头路了，我不可能放弃这么多年所领悟的东西去相信一批比我迷惘比我弱小的人

我姑且相信我只是年纪太轻不会装神秘，也分不清哪些东西是内圣的真理，分不清哪些东西是外王用的play的小玩具.唯一的方法是实践出真知,反馈调节.（或者说这个边界把控的不是特别好）

这下诗歌都不能锐评了,我哪里是喜欢古诗?秦后有什么可读的吗?取其次,都达到"通伦理"的标准的吗?何况现代诗.我当然不是全盘否定,可能算数的就那么几个,深入去读岂不是乱读一气?

我也不想讨论格式和情感的问题,这个问题讲了成千上万遍了,文学中的这个问题就如同数学中的勾股定理,伟大但是真的没必要再提了.
